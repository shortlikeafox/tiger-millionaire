```python
df_exotics = pd.read_csv('export_test.csv')
df = pd.read_csv('../data/kaggle_data/ufc-master.csv')
```

```python
df_exotics
```

```python
r_fighter_list = df_exotics['r_fighter'].tolist()
b_fighter_list = df_exotics['b_fighter'].tolist()
date_list = df_exotics['date'].tolist()
r_dec_odds_list = df_exotics['r_dec_odds'].tolist()
b_dec_odds_list = df_exotics['b_dec_odds'].tolist()
r_sub_odds_list = df_exotics['r_sub_odds'].tolist()
b_sub_odds_list = df_exotics['b_sub_odds'].tolist()
r_ko_odds_list = df_exotics['r_ko_odds'].tolist()
b_ko_odds_list = df_exotics['b_ko_odds'].tolist()
```

```python
df['r_dec_odds'] = ''
```

```python
for n in range(len(r_fighter_list)):
    df.loc[(df.R_fighter == r_fighter_list[n]) & (df.B_fighter == b_fighter_list[n]) & (df.date == date_list[n]), ['r_dec_odds', 'b_dec_odds', 'r_sub_odds', 'b_sub_odds' , 'r_ko_odds', 'b_ko_odds']] = (r_dec_odds_list[n]), (b_dec_odds_list[n]), (r_sub_odds_list[n]), (b_sub_odds_list[n]), (r_ko_odds_list[n]), (b_ko_odds_list[n])     
```

```python
df.to_csv('master_test.csv', index=False)
```
