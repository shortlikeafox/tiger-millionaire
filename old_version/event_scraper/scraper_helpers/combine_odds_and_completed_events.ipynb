```python
#OK..... Let's get 2 lists.  Red fighters, and blue fighters
red_fighter_list = df['R_fighter'].values
blue_fighter_list = df['B_fighter'].values
```

```python
red_odds_list = [None] * len(red_fighter_list)
blue_odds_list = [None] * len(red_fighter_list)
```

```python
for n in range(len(red_fighter_list)):
    #print(red_fighter_list[n])
    for r in all_rows:
        if str(r).find(str(red_fighter_list[n])) > -1:
            #print(r)
            #OK.... It looks like it's going to be the 2nd span....
            temp_odds = r.find_all('td')[2].find('span')
            if temp_odds is not None:
                temp_odds = temp_odds.get_text().strip()
            red_odds_list[n] = (temp_odds)
            break
        #If not let's look it up in the alias list
        else:
            for a in range(len(alias_alias_list)):
                if str(red_fighter_list[n]) == alias_fighter_list[a]:
                    if str(r).find(str(alias_alias_list[a])) > -1:
                        #print(r)
                        #OK.... It looks like it's going to be the 2nd span....
                        temp_odds = r.find_all('td')[2].find('span')
                        if temp_odds is not None:
                            temp_odds = temp_odds.get_text().strip()
                        red_odds_list[n] = (temp_odds)
                        break
                    
            
for n in range(len(blue_fighter_list)):
    #print(red_fighter_list[n])
    for r in all_rows:
        if str(r).find(str(blue_fighter_list[n])) > -1:
            #OK.... It looks like it's going to be the 2nd span....
            temp_odds = r.find_all('td')[2].find('span')
            if temp_odds is not None:
                temp_odds = temp_odds.get_text().strip()
            blue_odds_list[n] = (temp_odds)
            break
        else:
            for a in range(len(alias_alias_list)):
                if str(blue_fighter_list[n]) == alias_fighter_list[a]:
                    if str(r).find(str(alias_alias_list[a])) > -1:
                        #print(r)
                        #OK.... It looks like it's going to be the 2nd span....
                        temp_odds = r.find_all('td')[2].find('span')
                        if temp_odds is not None:
                            temp_odds = temp_odds.get_text().strip()
                        blue_odds_list[n] = (temp_odds)
                        break
        
```

```python
r_ev_list = []
b_ev_list = []
for r in red_odds_list:
    if r is not None:
        if int(r) > 0:
            r_ev_list.append(r)
        else:
            r_ev_list.append(-(100/int(r)) * 100)
    else:
        r_ev_list.append(None)
        
for b in blue_odds_list:
    if b is not None: 
        if int(b) > 0:
            b_ev_list.append(b)
        else:
            b_ev_list.append(-(100/int(b)) * 100)       
    else:
        b_ev_list.append(None)
```

```python
def return_odds(fighter, rows, phrase):
    l_name = fighter.split(' ')[-1]
    for r in rows:
        if str(r).find(l_name + ' ' + phrase) > -1:
            #It's the 2nd one.  Who knows why?
            all_cells = r.find_all('td')
            if len(all_cells) > 0:
                #2 is currently DraftKings,
                #10 is Ref
                if(len(all_cells[2].get_text().strip()))>2:
                    return(all_cells[2].get_text().strip().encode("ascii", "ignore").decode())
                elif(len(all_cells[10].get_text().strip()))>2:
                    return(all_cells[10].get_text().strip().encode("ascii", "ignore").decode())
                else:
                    return(all_cells[9].get_text().strip().encode("ascii", "ignore").decode())  
#       print(l_name)
        for a in range(len(alias_alias_list)):
            if fighter == alias_fighter_list[a]:
                #print(fighter, alias_fighter_list[a])
                l_name = alias_alias_list[a].split(' ')[-1]
                if str(r).find(l_name + ' ' + phrase) > -1:
                    if len(all_cells) > 0:
                        #2 is currently DraftKings,
                        #10 is Ref
                        if(len(all_cells[2].get_text().strip()))>2:
                            return(all_cells[2].get_text().strip().encode("ascii", "ignore").decode())
                        elif(len(all_cells[10].get_text().strip()))>2:
                            return(all_cells[10].get_text().strip().encode("ascii", "ignore").decode())
                        else:
                            return(all_cells[9].get_text().strip().encode("ascii", "ignore").decode())     
```

```python
while len(event_b_fighter_list) > 0:
    b_fighter = event_b_fighter_list.pop(0)
    r_fighter = event_r_fighter_list.pop(0)

    b_last_name = b_fighter.split(' ')[-1]
    r_last_name = r_fighter.split(' ')[-1]

    r_dec_odds = (return_odds(r_fighter, all_rows, 'wins by decision'))
    r_ko_odds = (return_odds(r_fighter, all_rows, 'wins by TKO/KO'))
    r_sub_odds = (return_odds(r_fighter, all_rows, 'wins by submission'))

    b_dec_odds = (return_odds(b_fighter, all_rows, 'wins by decision'))
    b_ko_odds = (return_odds(b_fighter, all_rows, 'wins by TKO/KO'))
    b_sub_odds = (return_odds(b_fighter, all_rows, 'wins by submission'))


    r_fighter_list.append(r_fighter)
    b_fighter_list.append(b_fighter)
    date_list.append(d)
    r_dec_list.append(r_dec_odds)
    r_ko_list.append(r_ko_odds)
    r_sub_list.append(r_sub_odds)
    b_dec_list.append(b_dec_odds)
    b_ko_list.append(b_ko_odds)
    b_sub_list.append(b_sub_odds)
```
